```python
import numpy as np

e = 10 ** (-6)

x = -3
fx = (x**3) - (9*x) + 3

it = 0
if np.abs(fx) < e:
  r = x
else:
  while np.abs(fx) > e:
    it += 1
    x0 = x
    x = ((x**3)/(9))+(1/3)
    fx = (x**3) - (9*x) + 3
    if (np.abs(fx) < e) or (np.abs(x-x0) < e):
      r = x
      break

print('Com %d iterações, a raiz é aproximadamente %s' %(it, r))
```

```python
import numpy as np

B = 0
NR = 0
S = 0

#--------------------------------------------------------------#
#Raízes pelo método da Bissecção

e = 10 ** (-7)

a = 0.5
b = 1
x0 = (a+b)/2

while (np.abs(a - np.tanh(1.5*a)) > e) or (np.abs(b - np.tanh(1.5*b)) > e):
  B += 1
  fa = a - np.tanh(1.5*a)
  fb = b - np.tanh(1.5*b)
  k = fa*fb
  if fa == 0:
    break
  elif fb == 0:
    break
  else: 
      if k < 0:
        fx0 = x0 - np.tanh(1.5*x0)
        if fx0 == 0 or (np.abs(fx0) < e):
          a = x0
          b = x0
          break
        else:
          if (fa > 0) and (fb < 0) and (fx0 >= 0):
            a = x0
            b = b
          elif (fa < 0) and (fb > 0) and (fx0 >= 0):
            a = a
            b = x0
          elif (fa > 0) and (fb < 0) and (fx0 <= 0):
            a = a
            b = x0
          elif (fa < 0) and (fb > 0) and (fx0 <= 0):
            a = x0
            b = b
      else:
        break
  x0 = (a+b)/2
if fa > fb:
  print('A quantidade de iterações que o método da Bissecção teve foi %d' %(B))
  print('A raiz é aproximadamente', b)
  print('\n')
elif fa < fb:
  print('A quantidade de iterações que o método da Bissecção teve foi %d' %(B))
  print('A raiz é aproximadamente', a)
  print('\n')
else:
  print('A quantidade de iterações que o método da Bissecção teve foi %d' %(B))
  print('A raiz é aproximadamente', a)
  print('\n')

#--------------------------------------------------------------#
#Raízes pelo método Newton-Raphson

e = 10 ** (-7)

x = 0.8
fx = x - np.tanh(1.5*x)

if np.abs(fx) < e:
  NR += 1
  r = x
else:
  while np.abs(fx) > e:
    NR += 1
    x0 = x
    x = x - ((x - np.tanh(1.5*x))/(1.5*(np.tanh(1.5*x))**2 - 0.5))
    fx = x - np.tanh(1.5*x)
    if (np.abs(fx) < e) or (np.abs(x-x0) < e):
      r = x
      break

print('A quantidade de iterações que o método Newton-Raphson teve foi %d' %(NR))
print('A raiz é aproximadamente', r)
print('\n')

#--------------------------------------------------------------#
#Raízes pelo método das Secantes

e = 10 ** (-7)

x0 = -2.4
x = -2.3
fx = np.tanh(1.5*x)

if np.abs(fx) < e:
  S += 1
  r = x
else:
  while np.abs(fx) > e:
    S += 1
    x1 = ((x0*(np.tanh(1.5*x)))-(x*(np.tanh(1.5*x0))))/((np.tanh(1.5*x))-(np.tanh(1.5*x0)))
    fx = np.tanh(1.5*x1)
    if (np.abs(fx) < e) or (np.abs(x1-x) < e):
      r = x1
      break
    x = x1
    x0 = x

print('A quantidade de iterações que o método das Secantes teve foi %d' %(S))
print('A raiz é aproximadamente', r)
```
